```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("cuentas_credicel.csv", encoding = 'latin')
df.head()
```

```python
##todas las columnas excepto "folio", "folio_solicitud", "Unnamed: 33", "Unnamed: 34", "Unnamed: 35", "Unnamed: 36"

df2 = df.drop(["folio", "folio_solicitud", "Unnamed: 33", "Unnamed: 34", "Unnamed: 35", "Unnamed: 36"], axis=1)
df2
```

```python
##todas las filas multiplos de 4
df3 = df2.iloc[::4, :]
df3
```

```python
##registros cuya marca sea igual a "alcatel motoraloa nokia zte y samsung"
df4 = df3[df3["marca"].isin(["ALCATEL", "MOTOROLA", "NOKIA", "ZTE", "SAMSUNG"])]
df4
```

```python
##Todos los registros cuyo enganche se encuentre entre 500 y 2000
##filtro7=df[(df["MONTO ADEUDADO"] < 6000) & (df["FECHA_FACTURA"] == "2023-03-14")] ejemplo

df5 = df4[(df4["enganche"] >= 500) & (df4["enganche"] <= 2000)]
df5["enganche"].unique()
```

```python
##Todos los registros cuya categoría en riesgo: NO sea “Cancelado” ni “Fraude”
##filtro9=df[~(df["MONTO ADEUDADO"] >= 6000)]

df6 = df4[~df4["riesgo"].isin(["Cancelado", "Fraude"])]
df6["riesgo"].unique()
print(df6)
```

```python
##f) Todos los registros cuya fecha : Sea posterior a “15/11/2021”
##
df7 = pd.to_datetime(df6["fecha"], errors="coerce", dayfirst=True)
df7 = df6[df6["fecha"] > "2021-11-15"]
df7.head()
```

```python
df7.to_csv("A7_filtros.csv", index=False, encoding="utf-8")
```

```python
##  a) Todas las columnas tipo fecha (datetime): Reemplazar nulos con el string: “01/01/2030 00:00”
##fecha_ultimo_pago fecha_pago_proximo fecha
df["fecha"] = df["fecha"].fillna("01/01/2030 00:00")
df["fecha_ultimo_pago"] = df["fecha_ultimo_pago"].fillna("01/01/2030 00:00")
df["fecha_pago_proximo"] = df["fecha_pago_proximo"].fillna("01/01/2030 00:00")

dff2 = df
dff2.head()
```

```python
##b todas las columnas tipo string object reemplazar nulos con el string "sin registro"
for columnas in dff2.select_dtypes(include="object").columns:
    dff2[columnas] = dff2[columnas].fillna("sin registro")

dff3 = dff2
dff3.head()
```

```python
##c todas las columnas tipo numerico float o int reemplazar nulos con el string sin registro
for columnas in dff3.select_dtypes(include=["float64", "int64"]).columns:
    dff3[columnas] = dff3[columnas].fillna(dff3[columnas].mean())

dff4 = dff3
dff4.head()
```

```python
dff4.to_csv("4_filtros.csv", index=False, encoding="utf-8")
```

```python
##metodos desviacion estandar
import numpy as np
import matplotlib.pyplot as plt
```

```python
dff5 = dff4.copy()
```

```python
###desviacion  estandar
#Limite_Superior = y.mean() + 3*y.std()
#Limite_Inferior = y.mean() - 3*y.std()
#print("Límite superior permitido:", Limite_Superior)
#print("Límite inferior permitido:", Limite_Inferior)
##


for columna in dff4.select_dtypes(include=["float64","int64"]).columns:
    media = dff4[columna].mean()
    desviacion = dff4[columna].std()
    
    Limite_Superior = media + 3*desviacion
    Limite_Inferior = media - 3*desviacion
    
    print("Columna:", columna)
    print("Límite superior permitido:", Limite_Superior)
    print("Límite inferior permitido:", Limite_Inferior)
```

```python
####los valores fuera de los límites se convierten en nulos

data3 = dff4.copy()

for columna in data3.select_dtypes(include=["float64","int64"]).columns:
    media = data3[columna].mean()
    desviacion = data3[columna].std()
    
    Limite_Superior = media + 3*desviacion
    Limite_Inferior = media - 3*desviacion
    
    data3[columna] = np.where((data3[columna] > Limite_Superior) | 
                              (data3[columna] < Limite_Inferior),
                              np.nan, 
                              data3[columna])

data3.head()
```

```python
##verificamos cuántos nulos quedaron después de eliminar outliers
##estoy siguiendo los pasos del ejercicio 5 valores atipicos

valores_nulos = data3.isnull().sum()
valores_nulos
```

```python
##reemplazamos los valores atípicos por el promedio de cada columna

data_clean = data3.copy()
data_clean = data_clean.fillna(data3.mean(numeric_only=True).round(1))
```

```python
#Corroboramos valores nulos del dataframe LIMPIO

valores_nulos2 = data_clean.isnull().sum()
valores_nulos2
```

```python
###Realizamos diagrama de caja o bigote de cada columna del dataframe
##fig = plt.figure(figsize =(15, 8))
##cuantitativas.plot(kind='box', vert=False)
#plt.title("Valores Atípicos del Dataframe")
##plt.show() #dibujamos el diagrama



for column in data_clean.select_dtypes(include=["float64","int64"]).columns:
    plt.figure(figsize=(6,3))
    data_clean.boxplot(column=[column])
    plt.title(f"{column} - Después de la sustitución de outliers")
    plt.show()
```
